```python
# Sub-Product and Sub-issue can be filled with other
data['Sub-product'].fillna('Other',inplace=True)
data['Sub-issue'].fillna('Other',inplace=True)

# The user might not have commented and hence , we can replace the null values as Not mentioned
data['Consumer complaint narrative'].fillna('Not mentioned',inplace=True)

# Company public response having na can we replaced with Not responded
data['Company public response'].fillna('Not responded',inplace=True)

# State and Zip code if not mentioned (na) can be replaced with other
data['State'].fillna('Other',inplace=True)
data['ZIP code'].fillna('Other',inplace=True)

# Tags hold values like servicemem , older American or both , in case of null replace it with Unknown
data['Tags'].fillna('Unknown',inplace=True)

# Consumer consent provided? is not provided or provided or withdrawn , so in case of null replace it with status not clear
data['Consumer consent provided?'].fillna('status not clear',inplace=True)

# company response to consumer -> only closed cases were provided , so the null cases might be the Not responded cases
data['Company response to consumer'].fillna('Not responded',inplace=True)

# Consumer disputed? -> holds yes , no values , so replace the null cases with 'Not sure'
data['Consumer disputed?'].fillna('Not sure',inplace=True)
```

2 Given an unsorted array of integers, find the length of the longest continuous
increasing subsequence (subarray).

```python
list_num = list(map(int,input().split()))
max_val,temp,n=1,1,len(list_num)
for i in range(1,n):
    if(list_num[i]>list_num[i-1]):
        temp+=1
        max_val=max(max_val,temp)
    elif(list_num[i]<=list_num[i-1]):
        temp=1
print(max_val)
```

3 Given a list of non negative integers, arrange them such that they form the largest
number.

```python
list_num = input().split()
n = len(list_num)
for i in range(0,n):
    for j in range(i+1,n):
        if(int(list_num[i]+list_num[j]) > int(list_num[j]+list_num[i])):
            list_num[i],list_num[j] = list_num[j],list_num[i]
print(''.join(list_num[::-1]))
```

4. Store all the &quot;servlet-name&quot;, and &quot;servlet-class&quot; to a csv file from the attached
sample_json.json file using Python.

```python
import json
import csv

with open("/content/drive/MyDrive/Colab_Notebooks/DT A1 sample_json.json",'r') as f:
    data=json.load(f)
csv_data = {'servlet-name':[],'servlet-class':[]}
for row in data['web-app']['servlet']:
    csv_data['servlet-name'].append(row['servlet-name'])
    csv_data['servlet-class'].append(row['servlet-class'])

print(csv_data)
rows = list(zip(csv_data['servlet-name'], csv_data['servlet-class']))
csv_rows = [dict(zip(csv_data.keys(), row)) for row in rows]

with open('/content/drive/MyDrive/Colab_Notebooks/servlets.csv', 'w', newline='') as csvfile:
    fieldnames = csv_data.keys()

    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()
    writer.writerows(csv_rows)
```
